```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split 
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, f1_score, precision_score, recall_score
le=LabelEncoder()
```

```python
for col in animal_data.columns:
    if(animal_data[col].dtype=='object'):
        animal_data[col]=le.fit_transform(animal_data[col])
```

```python
animal_data
```

```python
X=animal_data.drop('Dangerous',axis=1)
y=animal_data['Dangerous']

X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.25,random_state=0)
```

```python
model_svm=SVC(kernel='linear', random_state=0)
model_svm.fit(X_train,y_train)
```

```python
predicted_y=model_svm.predict(X_test)
print(predicted_y)
```

```python
Accuracy_score_svm=accuracy_score(y_test,predicted_y)
print('Accuracy: ',Accuracy_score_svm )
```

```python
f1_score_svm=f1_score(y_test,predicted_y)
precision_score_svm=precision_score(y_test,predicted_y)
recall_score_svm=recall_score(y_test,predicted_y)

print('f1_score_svm :',f1_score_svm)
print('precision_score_svm :',precision_score_svm)
print('recall_score_svm :',recall_score_svm)
```
